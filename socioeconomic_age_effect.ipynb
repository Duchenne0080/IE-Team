```
# import library
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```
"""
Source: Data Tables: Cardiovascular disease 2020 from Australian Institute of Health and Welfare
https://www.aihw.gov.au/reports/heart-stroke-vascular-diseases/cardiovascular-health-compendium/data
"""

# cause of hospitalisation for CVD
chd = ['Coronary Heart Disease', 'Other CVD']
hospitalisation = [161814,583923]
fig, ax = plt.subplots()
plt.pie(hospitalisation, labels = chd, autopct='%1.2f%%')
ax.set_title('Cause of Hospitalisation for \n Cardiovascular Disease 2017-18')
plt.show()

"""
Source: Household Expenditure and Characteristics, Victoria from ABS
6530.0 - Household Expenditure Survey, Australia: Summary of Results, 2015-16
https://www.abs.gov.au/AUSSTATS/abs@.nsf/DetailsPage/6530.02015-16?OpenDocument
"""

# Median household weekly gross income
income = ['508','1031','1626','2339','3827']
quintile = ['Lowest','Second','Third','Fourth','Highest']
df_inc = pd.DataFrame({'Income Quintile':quintile, 'Income ($)':income})

# Average household expenditure
## food and non-alcoholic beverages
food_beverage = ['118.16','167.59','224.38','311.55','396.73']
## medical care and health expenses
medical_health = ['37.83','53.29','71.82','89.72','138.67']

df_food_med = pd.DataFrame({'Income Quintile':quintile, 'Food & Beverage':food_beverage, 'Medical & Health':medical_health})
df_food_med = df_food_med.melt(id_vars=['Income Quintile'], value_vars=['Food & Beverage','Medical & Health'])
df_food_med = df_food_med.rename(columns={'variable':'Type', 'value':'Expense ($)'})

# plot data
fig, ax = plt.subplots(1,2, figsize=(12,5))
sns.barplot(x='Income Quintile', y='Income ($)', data=df_inc, ax=ax[0])
sns.barplot(x='Income Quintile', y='Expense ($)', hue='Type', data=df_food_med, ax=ax[1])
ax[0].set_title('Median Household Weekly Gross by Income')
ax[1].set_title('Average Household Expenditure by Income')

plt.tight_layout()
plt.plot()

"""
Source: Household Expenditure and Characteristics, Victoria from ABS
6530.0 - Household Expenditure Survey, Australia: Summary of Results, 2015-16
https://www.abs.gov.au/AUSSTATS/abs@.nsf/DetailsPage/6530.02015-16?OpenDocument
"""

# effect of socioeconomic on CVD hospitalisation
hos1 = [2227,2080,2005,1871,1826]
df_socio = pd.DataFrame({'Quintile':quintile, 'Hospitalisation (per 10e5)':hos1})

# effect of age on CHD hospitalisation
age = ['<25','25-34','35-44','45-54','55-64','65-74','75-84','85+']
hos2 = [1,20,134,517,1236,2331,3299,3068]
df_age = pd.DataFrame({'Age Group':age, 'Hospitalisation (per 10e5)':hos2})

# plot data
fig, ax = plt.subplots(1,2, figsize=(12,5))
sns.barplot(x='Quintile', y='Hospitalisation (per 10e5)', data=df_socio, ax=ax[0])
sns.barplot(x='Age Group', y='Hospitalisation (per 10e5)', data=df_age, ax=ax[1])
ax[0].set_title('CVD Hospitalisation by Socioeconomic Measures')
ax[1].set_title('CHD Hospitalisation by Age Groups')
plt.tight_layout()
plt.show()

"""
Source: Poor diet from Australian Institute of Health and Welfare
https://www.aihw.gov.au/reports/food-nutrition/poor-diet/data
"""

# fruit and vege intake by socioeconomic measures
no_fruit = [52.6,50.9,50.1,47.6,45.9]
no_vege = [94.3,92.5,93,91.9,91.6]
df_diet = pd.DataFrame({'Quintile':quintile, 'No Fruit':no_fruit, 'No Vege':no_vege})
# df_diet = df_diet.melt(id_vars=['Quintile'], value_vars=['No Fruit','No Vege'])
# df_diet = df_diet.rename(columns={'variable':'Diet', 'value':'Percentage (%)'})

fig, ax = plt.subplots(1,2, figsize=(12,5))
sns.barplot(x='Quintile', y='No Fruit', data=df_diet, ax=ax[0])
sns.barplot(x='Quintile', y='No Vege', data=df_diet, ax=ax[1])
fig.suptitle('Inadequate Diet by Socioeconomic Measures', y=1.05)
ax[0].set_title('Inadequate Fruits')
ax[1].set_title('Inadequate Vegetables')
plt.tight_layout()
plt.show()
```
